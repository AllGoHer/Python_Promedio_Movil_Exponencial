```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargar los datos desde el archivo CSV
```

```python
traficoweb = pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\traficoweb.csv', index_col = 'mes')
```

```python
traficoweb.head()
```

```python
# Asegurar que el índice esté en formato de fecha
```

```python
traficoweb.index = pd.to_datetime(traficoweb.index)
```

```python
# Calcular el Promedio Móvil Exponencial (PME) para una ventana de 6 meses
# Utiliza el método ewm() de Pandas para esto
# .ewm() es un método en Pandas que proporciona la funcionalidad para calcular un promedio móvil exponencial.
# El parámetro span=6 define el tamaño de la ventana para el promedio. En este caso, el promedio se calcula tomando
# en cuenta los últimos 6 períodos. A diferencia de un promedio móvil simple, en el promedio móvil exponencial,
# los valores más recientes tienen más peso, es decir, influyen más en el promedio.
# Esto hace que el PME sea más sensible a los cambios recientes en los datos.
```

```python
traficoweb['PME6'] = traficoweb['visitas'].ewm(span=6).mean()
```

```python
# Calcular el Promedio Móvil Exponencial (PME) para una ventana de 12 meses
# De nuevo, utiliza el método ewm()
```

```python
traficoweb['PME12'] = traficoweb['visitas'].ewm(span=12).mean()
```

```python
#visualizacion de datos
plt.figure(figsize=(12,6), dpi=300)
plt.plot(traficoweb['visitas'], label='Visitas Reales')
plt.plot(traficoweb['PME6'], label='PME 6 Meses')
plt.plot(traficoweb['PME12'], label='PME 12 Meses')
plt.title('Analisis del Trafico Web con Promedio Movil Exponencial')
plt.xlabel('Mes')
plt.ylabel('Visitas en  Miles')
plt.legend()
plt.show()
```
